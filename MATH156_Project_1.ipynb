```python
import random
import math
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('fitting_N20.csv', header=None)
```

```python
real_data = data.iloc[:,1:]
```

```python
real_data = real_data.reset_index(drop=True)
```

```python
x = real_data.iloc[:, 0]
t = real_data.iloc[:, 1]
```

```python
fullinterval = np.linspace(0, 2, num=100)


p2 = np.polyfit(x,t,4)  # fit a polynomial of order 2 to data points
curve2, = plt.plot(fullinterval, list(map(lambda x: np.polyval(p2, x), fullinterval)), 'g--', label="polynomial model of order 2")
point = plt.scatter(x,t, label="data points")
plt.legend(handles=[curve2,point], loc="upper right")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
p0 = np.polyfit(x,t,0)  # fit a polynomial of order 0 to data points
p1 = np.polyfit(x,t,1)  # fit a polynomial of order 1 to data points
p3 = np.polyfit(x,t,3)  # fit a polynomial of order 3 to data points
p5 = np.polyfit(x,t,5)  # fit a polynomial of order 3 to data points

p9 = np.polyfit(x,t,9)  # fit a polynomial of order 9 to data points
p12 = np.polyfit(x, t, 12) # fit a polynomial of order 12 to data points
p13 = np.polyfit(x, t, 13) # fit a polynomial of order 12 to data points
p14 = np.polyfit(x, t, 14) # fit a polynomial of order 12 to data points
p15 = np.polyfit(x, t, 15) # fit a polynomial of order 12 to data points

p20 = np.polyfit(x, t, 20) # fit a polynomial of order 20 to data points
p25 = np.polyfit(x, t, 25) # fit a polynomial of order 20 to data points
```

```python

trainerror = [0]*24
testerror = [0]*24
for i in range(1, 25):
    p = np.polyfit(x, t, i)  # compute polynomial
    trainpolyvals = np.polyval(p, x)  # of different orders
    trainerror[i-1] = math.sqrt(sum(list(map(lambda x: math.pow(x,2), trainpolyvals-t)))/20)
    testpolyvals = np.polyval(p, x)  # compute errors
    testerror[i-1] = math.sqrt(sum(list(map(lambda x: math.pow(x,2), testpolyvals-t)))/20)
    
# plot training and testing errors for different order models
curve1, = plt.plot(list(range(1,25)), trainerror, 'b', label="Training")
curve2, = plt.plot(list(range(1,25)), testerror, 'r', label="Testing")
plt.legend(handles=[curve1,curve2], loc="upper right")
plt.xlabel('Order')
plt.ylabel('Error')
plt.ylim((0,1))

plt.show()
```
